```python
tmp_data=[]
for i,d in data.iterrows():
    #uM/L --> mM/L
    if d['DIC_units']=='M/L':
        tmp_data.append(d['DIC']/1000)
    #mM/L --> mM/L
    elif d['DIC_units']=='mM/L':
        tmp_data.append(d['DIC'])
    #mM/Kg --> mM/L
    elif d['DIC_units']=='mM/kg':
        tmp_data.append(d['DIC'])
    #uM/kg --> mM/L
    elif d['DIC_units']=='M/kg':
        tmp_data.append(d['DIC']/1000) 
    #mg C/L --> mM/L
    elif d['DIC_units']=='mg C/L':
        tmp_data.append(d['DIC']/12.011)
    else:
        tmp_data.append(np.nan)
        
tmp_data=np.array(tmp_data)
data['DIC_transformed']=tmp_data.astype(float)
# data['DIC_transformed']==tmp_data['DIC_transformed'].astype(float)
#print(data['DIC_transformed'])
```

```python
Ca_data=[]
for i,d in data.iterrows():
    #uM/L --> mM/L
    if d['[Ca]_units']=='M/L':
        Ca_data.append(d['[Ca]']/1000)
    #mM/L --> mM/L
    elif d['[Ca]_units']=='mM/L':
        Ca_data.append(d['[Ca]'])
    #mg/L --> mM/L
    elif d['[Ca]_units']=='mg/L':
        Ca_data.append(d['[Ca]']/40.078)
    else:
        Ca_data.append(np.nan)
        
Ca_data=np.array(Ca_data)
data['Ca_trans']=Ca_data.astype(float)
```

```python
Mg_data=[]
for i,d in data.iterrows():
    #uM/L --> mM/L
    if d['[Ca]_units']=='M/L':
        Mg_data.append(d['[Mg]']/1000)
    #mM/L --> mM/L
    elif d['[Ca]_units']=='mM/L':
        Mg_data.append(d['[Mg]'])
    #mg/L --> mM/L
    elif d['[Ca]_units']=='mg/L':
        Mg_data.append(d['[Mg]']/24.305)
    else:
        Mg_data.append(np.nan)
        
Mg_data=np.array(Mg_data)
data['Mg_trans']=Mg_data.astype(float)
```

```python
#converting pCO2 units to atm
pCO2_data=[]
for i,d in data.iterrows():
    if d['CO2_units']=='atm':
        pCO2_data.append(d['CO2'])
    elif d['CO2_units']=='log(pCO2)':
        pCO2_data.append((10**(d['CO2']))*1000)
    elif d['CO2_units']=='log(pCO2)(mbar)':
        pCO2_data.append(((10**(d['CO2']))*0.000986923)*1000)
    elif d['CO2_units']=='mbar':
        pCO2_data.append((d['CO2']*0.000986923)*1000)
    elif d['CO2_units']=='Pa':
        pCO2_data.append((d['CO2']*9.8692300004935e-6)*1000)
#convert here (coursepack)
    elif d['CO2_units']=='ppm':
        pCO2_data.append(np.nan)
    elif d['CO2_units']=='ppmv':
        pCO2_data.append(np.nan)            
    elif d['CO2_units']=='M/L':
        pCO2_data.append(np.nan)
    else:
        pCO2_data.append(np.nan)
        
pCO2_data=np.array(pCO2_data)
data['pCO2_trans']=pCO2_data.astype(float)
```

```python
x = data['pCO2_trans']
y = data['d13C_DIC']
plt.figure(figsize=(10,8))
plt.scatter(x,y)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.ylabel('d13C_DIC (ppm)',fontsize=18)
plt.xlabel('pCO2 (atm)',fontsize=18)
plt.title('d13C_DIC vs. pCO2', fontsize=20)
plt.xlim(0,2000)
```
