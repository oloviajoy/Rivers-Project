```python
#tool box
import pandas as pd
from scipy import stats
import numpy as np
from matplotlib import pyplot as plt

#stylin'
plt.style.use('seaborn-darkgrid')

#import data
data = pd.read_excel('filtered_lat_long.xlsx')
```

```python
#latitude vs d13C...
y = data['lat']
x = data['d13C_DIC']

plt.figure(figsize=(10,8))
plt.scatter(x,y)
plt.ylabel('Latitude',fontsize=18)
plt.xlabel('$\delta^{13}$C',fontsize=18)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.grid(color='lightgrey')
```

```python
#latitude vs d13C...
y = data['lat']
x = data['DIC_trans']*1000

plt.figure(figsize=(10,8))
plt.scatter(x,y)
plt.ylabel('Latitude',fontsize=18)
plt.xlabel('DIC',fontsize=18)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.grid(color='lightgrey')
```

```python
#latitude vs d13C...
y = data['lat']
x = data['d13C_DIC']

plt.figure(figsize=(10,12))
plt.scatter(x,y)
plt.ylabel('Latitude',fontsize=18)
plt.xlabel('$\delta^{13}$C',fontsize=18)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.title('Latitude vs. $\delta^{13}$C', fontsize=20)
```

```python
import matplotlib.patches as patches
fig, ax = plt.subplots(figsize=(12,12))
rect_Com = patches.Rectangle((0, -1), 3000, 2, edgecolor='none', facecolor='g',alpha=.3)
rect_Ccw = patches.Rectangle((0, -11.5), 4000, 6, edgecolor='none', facecolor='orange',alpha=.3)
rect_Cncw = patches.Rectangle((0, -23), 2000, 8, edgecolor='none', facecolor='b',alpha=.3)
ax.add_patch(rect_Com)
ax.add_patch(rect_Ccw)
ax.add_patch(rect_Cncw)

x=data['DIC_trans']*1000
plt.scatter(x,data['d13C_DIC'])



plt.ylabel('$\delta^{13}$C$_{\mathrm{DIC}}$ ',fontsize=18)
plt.xlabel('DIC (M/L)',fontsize=18)
plt.title('$\delta^{13}$C$_{\mathrm{DIC}}$ vs. DIC', fontsize=20)
plt.legend(loc='best',fontsize='x-large',frameon=False)
```

```python
#plot d13C vs DIC with latitude ranges...
idx = data['lat']>=40
idx2 = (data['lat']>=20) & (data['lat']<40)
idx3 = (data['lat']>=-10) & (data['lat']<20)
idx4 = (data['lat']>=-20) & (data['lat']<-10)
idx5 = (data['lat']>=-40) & (data['lat']<-20)
```

```python
import matplotlib.patches as patches
fig, ax = plt.subplots(figsize=(12,12))
rect_Com = patches.Rectangle((0, -1), 3000, 2, edgecolor='none', facecolor='g',alpha=.3)
rect_Ccw = patches.Rectangle((0, -11.5), 4000, 6, edgecolor='none', facecolor='orange',alpha=.3)
rect_Cncw = patches.Rectangle((0, -23), 2000, 8, edgecolor='none', facecolor='b',alpha=.3)
ax.add_patch(rect_Com)
ax.add_patch(rect_Ccw)
ax.add_patch(rect_Cncw)

idx = data['lat']>=40
x=data['DIC_trans'][idx]*1000
plt.scatter(x,data['d13C_DIC'][idx],label='40+')

idx = (data['lat']>=20) & (data['lat']<40)
x=data['DIC_trans'][idx]*1000
plt.scatter(x,data['d13C_DIC'][idx],label='20 to 40')

x=data['DIC_trans'][idx3]*1000
plt.scatter(x,data['d13C_DIC'][idx3],label='-10 to 20')

x=data['DIC_trans'][idx4]*1000
plt.scatter(x,data['d13C_DIC'][idx4],label='-20 to -10')

x=data['DIC_trans'][idx4]*1000
plt.scatter(x,data['d13C_DIC'][idx4],label='-40 to -20')


plt.ylabel('$\delta^{13}$C$_{\mathrm{DIC}}$ ',fontsize=18)
plt.xlabel('DIC (M/L)',fontsize=18)
plt.title('$\delta^{13}$C$_{\mathrm{DIC}}$ vs. DIC', fontsize=20)
plt.legend(loc='best',fontsize='x-large',frameon=False)
```
