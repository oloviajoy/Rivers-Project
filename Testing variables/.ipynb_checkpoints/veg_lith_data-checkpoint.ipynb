```python
#x-d13C, y-veg
#can categorize- set boundaries 0.1-0.25 = desert or whatever.
fig, ax = plt.subplots(figsize=(12,8))
plt.scatter(data['veg_value'],data['d13C_DIC'])

min_value=np.min(data['veg_value'])
max_value=np.max(data['veg_value'])

#add binned data mean range and standard deviation
bins=np.hstack((np.linspace(min_value,max_value,12)[0:-1].reshape(-1,1),np.linspace(min_value,max_value,12)[1:].reshape(-1,1)))
my_line=[]

for i,b in enumerate(bins):
    mu=np.mean(data['d13C_DIC'][(data['veg_value']>=b[0]) & (data['veg_value']<b[1])])
    std=np.std(data['d13C_DIC'][(data['veg_value']>=b[0]) & (data['veg_value']<b[1])])
    plt.plot([b[0],b[1]],[mu,mu],'g-') #x-range

    plt.plot([np.mean(b),np.mean(b)],[mu+std,mu-std],'r-')
    
    my_line.append([np.mean(b),mu])

my_line=np.array(my_line)
plt.plot(my_line[:,0],my_line[:,1], linestyle='dashed',color='black')

plt.ylabel('$\delta^{13}$C$_{\mathrm{DIC}}$ ‰',fontsize=18)
plt.xlabel('vegetation index',fontsize=18)
plt.title('$\delta^{13}$C$_{\mathrm{DIC}}$ vs. vegetation', fontsize=20)
plt.legend(loc='best',fontsize='x-large',frameon=False)
```
