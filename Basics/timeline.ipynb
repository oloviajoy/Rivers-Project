```python
##### tool box
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

#stylin'
plt.style.use('default')

#import the data
data=pd.read_csv('d13c_compilation.csv')
```

```python
x=data['age']
y=data['d13c']

plt.figure(figsize=(15,8))
plt.scatter(x,y,color='darkcyan')
plt.xlim(885,-3)
plt.ylim(-18,12)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.ylabel('$\delta^{13}$C$_{\mathrm{carb}}$ ()',fontsize=18)
plt.xlabel('Time (Ma)',fontsize=18)
```
